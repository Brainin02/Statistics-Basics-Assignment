5. Discuss the role of random sampling in making inferences about populations.

ans. The Role of Random Sampling in Making Inferences About Populations

a). What is Random Sampling?
Random sampling is a method of selecting individuals from a population such that each member has an equal chance of being chosen. It is a fundamental technique in statistical inference, allowing researchers to draw conclusions about an entire population based on a smaller subset (sample).

b).Why is Random Sampling Important?
âœ… Ensures Representativeness

A well-chosen random sample reflects the characteristics of the population, reducing bias.

Example: A nationwide political survey should include people of different ages, genders, and regions.

 Reduces Bias

Avoids selection bias, where certain groups are overrepresented or underrepresented.

Example: A medical study should not only test a drug on young individuals if itâ€™s meant for all age groups.

 Enables Generalization

Proper random sampling allows researchers to generalize findings from the sample to the larger population with a known margin of error.

Example: If a randomly selected group of 1,000 voters supports a candidate, we can estimate overall voter support.

 Allows for Statistical Inference

Techniques like confidence intervals and hypothesis testing rely on random sampling to estimate population parameters (e.g., mean, proportion).

 c). Types of Random Sampling

   Simple Random Sampling (SRS)

Every member has an equal chance of selection.

Example: Drawing names from a hat or using a random number generator.

 Stratified Random Sampling

Population is divided into subgroups (strata), and random samples are taken from each.

Example: Surveying 100 students by randomly selecting 25 from each grade level.

 Systematic Sampling

Selecting every kth individual after a random starting point.

Example: Choosing every 10th customer in a store.

 Cluster Sampling

Population is divided into groups (clusters), and entire clusters are randomly selected.

Example: Randomly selecting five schools in a district and surveying all students in those schools.


 d)..Limitations & Challenges of Random Sampling

 Sampling Errors

Random samples may still differ from the population, leading to variability (margin of error).

 Nonresponse Bias

If selected individuals refuse to participate, results may not fully represent the population.

 Difficulties in Implementation

True randomness requires careful planning, resources, and access to a full population list.

6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

ans. Skewness measures the asymmetry of a probability distribution relative to its mean. If a dataset is perfectly symmetrical, it has zero skewness. However, when data is concentrated on one side, the distribution is skewed either positively or negatively.

Types of Skewness

. Positive Skewness (Right-Skewed Distribution)

The right tail (higher values) is longer than the left tail.

The mean > median > mode (pulled right by higher values).

Example: Income distribution (a few high earners pull the average up).

 . Negative Skewness (Left-Skewed Distribution)

The left tail (lower values) is longer than the right tail.

The mean < median < mode (pulled left by lower values).

Example: Test scores where most students perform well, but a few get very low scores.

. Zero Skewness (Symmetrical Distribution)

The left and right sides are mirror images.

The mean = median = mode.

Example: A normal distribution (e.g., IQ scores).

How Skewness Affects Data Interpretation

 Influences Measures of Central Tendency

In a right-skewed distribution, the mean is higher than the median (due to large values).

In a left-skewed distribution, the mean is lower than the median.

The median is often a better measure of central tendency for skewed data.

 Impacts Statistical Analysis

Many statistical models assume normality (zero skewness); heavily skewed data may require transformation (e.g., log transformation).

 Affects Decision-Making

Right-skewed: When analyzing salaries, the median may give a more accurate picture than the mean.

Left-skewed: In analyzing test scores, a lower mean may not indicate poor performance if most students did well.
